```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
import urllib.request
import re
```

```python
# Go to the page all-cities in the Western Cape and load all of those links into a file
# Check if the city referred to in the link is in the ct_areas list
# If it is in the CT_areas list, then go into the page and go to the "All areas"
# Store each of the links in the "All areas" part of the page

# When it comes to storing a name, if the name has "Central" as the last word, remove "Central", this is to help it match up
# with the names in the Cape Town suburbs shapefile used in R

# https://www.property24.com/for-sale/all-cities/western-cape/9
```

```python
# Load the file that indicated
with open("CTAreas.csv") as ffile:
    ct_areas = ffile.read()
ct_areas = ct_areas.lower().split(' \n')
#ct_areas
```

```python
# Create a list of links to scrape from the page where you need to go into the links and scrape their links i.e. the links
# for 3) above
response = requests.get("https://www.property24.com/for-sale/all-cities/western-cape/9")
soup = BeautifulSoup(response.text, 'html.parser')
big_links_list = soup.findAll('a')
```

```python
# Create a list to capture all of the relevant suburb-specific links and then loop through all of the links on the
# parent "Western Cape" page on property24.com - this page has a more cities than the Western Cape page in 2) above
def get_links(links, sub_cities = False):
    # links - BeautifulSoup object (webpage) containing links to suburbs
    # Returns - a list of the relevant links from that page
    return_list = []
    for link in links:
        address = link.get('href')
        
        try:
            # This filters out the links so we're only looking at suburbs links (and some houses, which are filtered out)
            # Check that the link starts with /for-sale and ends with 4 or 5 numbers
            if re.search('^/for-sale', address) and re.search('[0-9]{3,4}$', address):
                temp_address = address

                # Condition to check whether we're pulling from the main page or the "cities" pages
                if sub_cities == True:
                    if address.split('/')[2] == "all-suburbs":
                        return_list.append(address)
                else:
                # This removes links to individual properties by checking whether the address ends with 6 numbers
                    if(len(temp_address.split('/')[-1])) < 6:
                        # This only adds the links for "cities" in Cape Town that in the pre-selected areas list
                        area = temp_address.split('/')[2]
                        area = re.sub("-", " ", area)
                        if area in ct_areas:
                            return_list.append(address)
        except:
            continue
    # Return a list of the unique links on the pages searched
    return_list = list(set(return_list))
    return(return_list)
```

```python
# Go into the link for each "city" in Cape Town and go to the link for "All suburbs in [Suburb]" which contains
# "all-suburbs" at [2] in the split by "/"
cities_links = get_links(big_links_list)

suburbs_links = get_links(cities_links)
```

```python
all_suburbs = []
for link in cities_links:
    response = requests.get("https://www.property24.com"+str(link))
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.findAll('a')
    all_suburbs.append(get_links(links, sub_cities = True))
all_suburbs = [item for sublist in all_suburbs for item in sublist]      
```

```python
# Cycle through all of the "Suburb main pages" and extract the links for each of the suburbs within
every_suburb = []
for link in all_suburbs:
    response = requests.get("https://www.property24.com"+str(link))
    soup = BeautifulSoup(response.text, 'html.parser')
    links = soup.findAll('a')
    every_suburbs.append(get_links(links, sub_cities = False)) 
```
